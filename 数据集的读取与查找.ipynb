# CSV介绍及读取CSV

能够手动创建 DataFrame 或 Series 非常方便。但是，大多数时候，我们实际上并不是手工创建我们自己的数据。相反，我们将使用已经存在的数据。

数据可以以多种不同的形式和格式存储。到目前为止，其中最基本的是 CSV 文件。当你打开一个 CSV 文件，你会看到这样的东西:

所以一个 CSV 文件是一个由逗号分隔的值表，因此它的名字是“逗号分隔值”，或者 CSV。

现在，让我们把数据集放在一边，看看将它读入 DataFrame 时，真正的数据集是什么样子的。我们将使用 pd.read _ csv ()函数将数据读入 DataFrame:

```python
import pandas as pd
wine_reviews = pd.read_csv("C://Users//qiao//Downloads//submit_result.csv",index_col=0)
```

在这里，我们通过"index_col = 0"来去掉不必要的索引;我们可以使用shape函数来查看读入的csv文件有几行几列

```python
wine_reviews.shape
```

由shape函数得知该表格共有11424行，34列，我们接下来看看该表格的头五行和尾五行

使用head函数和tail函数来查看前5行、末5行

```python
wine_reviews.head()
```

```python
wine_reviews.tail()
```

```python
wine_reviews
```

索引查找某具体的一列

```python
wine_reviews['上部温度6']
```

我们完全可以把DataFrame当作一个二维数组来进行查找,例如查找第六列第356行的值：

```python
wine_reviews['上部温度6'][356]
```

接下来介绍行索引：iloc函数

```python
wine_reviews.iloc[0]
```

使用分片来进行索引。例如查看所有行的第0列数据：

```python
wine_reviews.iloc[:,0]
```

查看前3行的第一列

```python
wine_reviews.iloc[:3, 0]
```

还有另外几种表示方法：

```python
wine_reviews.iloc[[0, 1, 2], 0]
```

```python
wine_reviews.loc[1,'上部温度2']#注意这里不能使用iloc函数
```

iloc 在概念上比 loc 简单，因为它忽略了数据集的索引。当我们使用 iloc 时，我们把数据集看作一个大矩阵(列表列表) ，我们必须按位置索引它。相比之下，loc 使用索引中的信息来完成它的工作。因为数据集通常具有有意义的索引，所以使用 loc 代替它通常更容易。
